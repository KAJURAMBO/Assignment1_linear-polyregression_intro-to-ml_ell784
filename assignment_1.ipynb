```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import mean_squared_error
import math
data=pd.read_csv('train.csv')
X=data['x']
Y=data['y']
X_20=X.loc[0:19]
Y_20=Y.loc[0:19]
# X_20.shape## As the dimension is 100, but linear regression trainning need atleast 1 dimension
# Y_20.shape
# X_20.ndim
# Y_20.ndim ## 1 ...Cant train linear regression model so we need to give dimension
x_train = X_20.values.reshape(-1,1)
y_train = Y_20.values.reshape(-1,1)

M=20
Mean_square_error=[]

for i in range(M):
        polynomial_features = PolynomialFeatures(degree = i)
        X_poly = polynomial_features.fit_transform(x_train)
        polynomial_reg_model=LinearRegression()
        polynomial_reg_model.fit(X_poly,y_train)
        y_predicted = polynomial_reg_model.predict(X_poly)
        y_predicted
        plt.figure(figsize=(5,5))
        plt.title(f"Graph for M ={i}")
        plt.scatter(X_20,Y_20)
        plt.xlabel("X")
        plt.ylabel("Y")
        plt.scatter(X_20,y_predicted,c="red")
        plt.show
        MSE = mean_squared_error(Y_20, y_predicted)
        RMSE = math.sqrt(MSE)
        Mean_square_error.append(RMSE)
        print("Mean Square Error:",RMSE)
Iterations_M= (np.linspace(1,20,num=20,dtype="int"))
plt.figure(figsize = (12, 5))
plt.plot(Iterations_M, Mean_square_error, 'green')
plt.title("Error vs M")
plt.xlabel('Iterations_M')
plt.ylabel('RMSE')
plt.xticks(np.arange(1,21))
plt.grid(True, color = "grey", linewidth = "1.4", linestyle = "-.") 
plt.show()
```
